## Seaborn Histogram

### In the above distribution,

- min number of age si 21 and 22 with more number of male and no female
- It is right skewed 
- Highest count is 58

```python
sns.displot(data=df, x="sex", y="age")
```

### The above figure is showing the histogram between sex and age
- The above distribution is between sex and age
- X-axis represents sex
- Y-axis represents age
- In age, the record is between 15 and 22 and in sex there is F and M
- There are maximum number of male and minimum number of female where age of 21 and 22 only exists in male group

## Seaborn Boxplot

```python
sns.boxplot(y = 'age',data= df, x= 'sex')
plt.xlabel('SEX')
```

### In the above boxplot between age and sex
- it is categorized as F and M
- median value in F is 17 and there is no median value in M
- the median in boxplot F is at Q2 so it has no skewness, and since there is no median in boxplot M, the median is equal to either Q1 or Q2 which means it has skewness
- In boxplot F, the min valur is 16 and the max value is 18, and alos in the boxplot M, the min value is 16 and max value is 18
- Boxplot F is normal but M is either right skewed or left skewed
- Outliers doesn't exist in F but exists in M since there is age outside the range of boxplot

```python
# This is heat map pearson correction matrix
corrmat = df.corr()
f, ax= plt.subplots(figsize=(12,10))
sns.heatmap(corrmat, vmax=.8, square=True)
```

### The above figure is about the heatmap pearson correction matrix
- Light color shown in right scale +0.8 is highly correlated and darker color below -0.2 is not correlated 
- Each square shows the correlation between the variables on each axis. 
- Correlation ranges from -0.2 to +0.8. 
- Values closer to zero means there is no linear trend between the two variables. 

- The closer to +0.8 the correlation is the more positively correlated they are, which means that as one increases so does the other and the closer to +0.8 the stronger this relationship is. 
- A correlation closer to -0.2 is similar, but instead of both increasing one variable will decrease as the other increases. 

- The diagonals are all white because thos/e squares are correlating each variable to itself (so it's a perfect correlation). 


### If it feels hard to look the color and see which one is highly correlated, we can do another style as shown below

```python
plt.figure(figsize=(20,20))
plt.title('Pearson Correlation of Features',size=15)
colormap= sns.diverging_palette(10,220, as_cmap=True)
sns.heatmap(df.corr(), cmap=colormap, square=True, annot=True, 
            linewidths=0.1, vmax=1.0, linecolor='white', annot_kws={'fontsize':12})
plt.show()
```

- In above correlation matrix, we printed the number also so it will be easy for us to see which are highly corelated
- Value close to 1.00 is highly correlated
